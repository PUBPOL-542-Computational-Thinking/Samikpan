The following section details the steps for cleaning data to prepare the dataset for analysis. 



To do the preparation, it is first important to import the function panda so it can read a data frame.Since the dataset is in excel file,after importing the panda,the function read excel must be called to ensure that python recognizes files that are in  the excel format. 


```python
import pandas as pd

location='https://github.com/Samikpan/PUBPOL542/raw/main/'
Financial_data=location+'Financial_Development_FD.xlsx'
Financial_index = pd.read_excel(Financial_data,'Financial Development (FD)')
```

Once the data file is read, copy of dataset is created so that the original dataset remains intact. 
Creaed copy is now renamed to Financial_IndexTest. All the cleanig data will occur to this copied dataset. 

To clean the dataset, it is important to first understand it. So to understand the format of data, using the code describe(). Now to understand the data type, I use code info(). Understadning the data type is important to proceed forweard with cleaning. 

```python
Financial_indexTest=Financial_index.copy()
```

```python
Financial_indexTest.describe()
```

```python
Financial_indexTest.info()
```

```python
Financial_indexTest[:]
```

After knowing the data type, I decided that to only keep the data column that are valuable for analysis. So,the next stop is to drop column that are not essential for further steps. After reading the data column, I created a group of columns I want to drop and used the function drop() to remove all the columns I will not be using in the data analysis. The code to do so can be observed in the code below.

```python
cols = [2,4,6,8,10,12,14,16,18]
Financial_indexTest.drop(Financial_index.columns[cols],axis=1,inplace=True)
```

After dropping the columns I dont need, I wanted to see the columns remaning so I used to function [:] This lists all the element in an array.  Since the first column, the name of the country is currently under the name "unnamed", I decided to rename it to "Country". I used the function rename() to replace the column name from unnamed to country. 

```python
Financial_indexTest[:]
```

```python
Financial_indexTest.rename(columns={'Unnamed: 0':'Country'}, inplace = True) ### renaming an unnamed column
```

After renaming the column to country, I realized that other column names are long. To make the analysis process simpler, I decided to shorten the name. For example I renamed the column name "Financial Development Index to FDI. I did the similar thing to rename Financial Institutions Effiiciency Index to "FIEI". I renamed the column using the position of column because the names of the variables are long. The function used is rename with column [] with the position I want to change them. 

```python
Financial_indexTest.rename(columns={Financial_indexTest.columns[1]:"FDI",Financial_indexTest.columns[4]: "FIEI"}, inplace = True)
```

```python
Financial_indexTest[:]
```

```python
Financial_indexTest.info()
```

After renaming the column,I decided to keep only 3 columns, Country, FDI and FIEI in the data analysis. After cleaning data, I will need to merge it with other dataset so it is important to retain only the information I need for anlysis.
To do this, I create a different file called "Financial_FinalData" that retains only column 0,1,4. Column 0 is included because the index in Python starts with O, therefore, the position of first column, i.e "Country" is O

```python
Financial_FinalData = Financial_indexTest.iloc[:, [0,1,4]]
```

After saving the final dataset htat consists of cleaned data,I will needto export it to excel csv format so that they can be merged with other dataset.The code utilized to do si is "to_Csv). 

```python
Financial_FinalData
```

```python
Financial_FinalData.to_csv('Samikdata.csv',index= False)
```
